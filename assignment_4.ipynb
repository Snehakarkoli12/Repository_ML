# 36.Is there any seasonality in the CPI trends?

```python

from statsmodels.tsa.seasonal import seasonal_decompose

df["Month & Year"] = pd.to_datetime(df["Month & Year"], format="%B, %Y")

df = df.sort_values("Month & Year")
df.set_index("Month & Year", inplace=True)

df.fillna(method="ffill", inplace=True)

if df.shape[0] >= 12:
    # Apply seasonal decomposition
    decomposition = seasonal_decompose(df["Combined"], model="additive", period=12)
    
    # Plot decomposition
    decomposition.plot()
    plt.show()
else:
    print("Not enough data for seasonal decomposition (Requires at least 12 months).")
```

# 37.What is the overall difference between Rural and Urban CPI?

# 59.  Can we use Linear Regression to predict next month's CPI?

```python
from sklearn.linear_model import LinearRegression
import numpy as np

df["Month_Index"] = np.arange(len(df))  

X = df[["Month_Index"]]
y = df["Combined"]

model = LinearRegression()
model.fit(X, y)

next_month_index = np.array([[len(df)]]) 
predicted_cpi = model.predict(next_month_index)
print(f"Predicted CPI for next month: {predicted_cpi[0]:.2f}")
```
